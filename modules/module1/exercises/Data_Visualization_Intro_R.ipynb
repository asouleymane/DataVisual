pairs of variables that are correlated are

```r
library(ggplot2)
library(gridExtra)
```

**Activity 4: ** Fit a linear regression model between Calories and TotalFat.

```r
## Your answer for activity 4 goes here ---
lm(Calories~TotalFat,USDA)
```

**Activity 5: ** Plot a grid of scatter plots for the 4 pairs of correlations identified in activity 3.

```r
## Your answer for activity 5 goes here ---
p1<-ggplot(USDA,aes(Calories,TotalFat)) + geom_point()
p1<-p1 + geom_smooth(method = lm, se = FALSE) + expand_limits(x = 800, y = 100)
p2<-ggplot(USDA,aes(Calories,SaturatedFat)) + geom_point()
p2<-p2 + geom_smooth(method = lm, se = FALSE) + expand_limits(x = 800, y = 100)
p3<-ggplot(USDA,aes(TotalFat,SaturatedFat)) + geom_point()
p3<-p3 + geom_smooth(method = lm, se = FALSE) + expand_limits(x = 100, y = 100)
p4<-ggplot(USDA,aes(Sugar,Carbohydrate)) + geom_point()
p4<-p4 + geom_smooth(method=lm, se = FALSE) + expand_limits(x = 100, y = 100)
grid.arrange(p1,p2,p3,p4)
```

**Activity 6: ** Plot a scatter plot between Calories and TotalFat. Plot the points in green color and transparent. Adjust the breaks for x and y axes. 

```r
## Your answer for activity 6 goes here ---

pp1<-ggplot(USDA,aes(Calories,TotalFat)) + geom_point(color="green",alpha = .05)
pp1<-pp1 + scale_x_continuous(breaks=seq(0,1000,250)) + scale_y_continuous(breaks=seq(0,100,25))
pp1<-pp1 + geom_smooth(method = lm, se = FALSE)
pp1
```
